```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from statsmodels.iolib.summary2 import summary_col
```

```python
n=100000
df=pd.DataFrame(index=range(n))
df['ability']=np.random.normal(100,15,n)
df['training']=(df.ability+ np.random.normal(0,10,n)>=100)*1
df['sales']=10000+df.training*5000+df.ability*100+np.random.normal(0,4000,n)
df['training2']= np.random.binomial(1,0.05,n)
df['sales']=df.sales+df.training2*5000
```

```python
reg1=smf.ols('sales ~ training', data=df).fit()
reg2=smf.ols('sales ~ training + ability', data=df).fit()

print(summary_col([reg1, reg2], regressor_order=['training', 'ability']))
```

```python
reg3=smf.ols('ability ~ training', data=df).fit()
print(reg3.summary())
```
