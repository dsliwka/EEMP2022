```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from statsmodels.iolib.summary2 import summary_col
```

```python
n=10000
df=pd.DataFrame(index=range(n))
df['ability']=np.random.normal(100,15,n)
df['training']=(df.ability+ np.random.normal(0,10,n)>=100)*1
df['sales']=10000+df.training*5000+df.ability*100+np.random.normal(0,4000,n)
```

```python
reg=smf.ols('sales ~ training', data=df).fit()
print(summary_col([reg], stars=True))
```

```python
df.groupby('training').ability.mean()
```

```python
reg2=smf.ols('sales ~ training + ability', data=df).fit()
print(summary_col([reg, reg2], regressor_order=['training', 'ability'], stars=True))
```

```python
df['testScore'] = 100 + df['ability'] + df['training']*50  + np.random.normal(0,5,n)

reg1=smf.ols('sales ~ training', data=df).fit()
reg2=smf.ols('sales ~ training + testScore', data=df).fit()
print(summary_col([reg1, reg2], regressor_order=['training', 'ability', 'testScore'], stars=True))
```

```python
df['hiringTest']=df.ability+np.random.normal(0,8,n)
reg1=smf.ols('sales ~ ability', data=df).fit()
reg2=smf.ols('sales ~ hiringTest', data=df).fit()
print(summary_col([reg1, reg2], regressor_order=['ability', 'hiringTest'], stars=True))
```

```python
reg1=smf.ols('sales ~ training', data=df).fit()
reg2=smf.ols('sales ~ training + hiringTest', data=df).fit()
reg3=smf.ols('sales ~ training + ability', data=df).fit()
print(summary_col([reg1, reg2, reg3], regressor_order=['training', 'hiringTest', 'ability'], stars=True))
```
